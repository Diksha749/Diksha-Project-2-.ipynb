**Q. Discuss string slicing and provide example.**

Ans. String Slicing in Python:
String slicing is extracting a portion of a string by specifying its start and end indices.

Syntax:
string[start:end:step]>>

* start: Index of the first character to include (default is 0).
* end: Index of the first character to exclude (default is length of the string).
* step: The interval between characters (default is 1).
Examples:

```python
Python
text = "Hello, World!"

# Extract "Hello"
print(text[0:5])  # Output: Hello

# Extract "World"
print(text[7:12])  # Output: World

# Extract every other character
print(text[::2])  # Output: Hlo ol!

# Reverse the string
print(text[::-1])  # Output: !dlroW ,olleH
```

Q. Describe how to access,modify,and delete elements in a list with examples.

Ans. Accessing, Modifying, and Deleting List Elements
Accessing Elements
Indexing: Use square brackets with the index to access an element.

```python
my_list = [10, 20, 30, 40]
print(my_list[0])  # Output: 10 (first element)
print(my_list[-1])  # Output: 40 (last element)
```

Slicing: Access a range of elements using [start:end:step].

```python
print(my_list[1:3])  # Output: [20, 30]
```

Modifying Elements
Direct assignment: Assign a new value to an element.


```python
my_list[2] = 50
print(my_list)  # Output: [10, 20, 50, 40]
```

Deleting Elements
del keyword: Delete an element by its index.

```python
del my_list[1]
print(my_list)  # Output: [10, 50, 40]
```

remove() method: Remove the first occurrence of a value.

```python
my_list.remove(50)
print(my_list)  # Output: [10, 40]
```

pop() method: Remove and return the element at a specified index (default is last).

```python
last_element = my_list.pop()
print(my_list)  # Output: [10]
print(last_element)  # Output: 40
```

**Q. Describe how to add, modify,and delete items in a dictionary with examples.**

Ans. Adding, Modifying, and Deleting Items in a Dictionary
Adding Items
Direct assignment:


```python
my_dict = {'name': 'Alice', 'age': 30}
my_dict['city'] = 'New York'  # Add a new key-value pair
print(my_dict)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
```

Modifying Items
Direct assignment:

```python
my_dict['age'] = 31  # Modify the value of an existing key
print(my_dict)  # Output: {'name': 'Alice', 'age': 31, 'city': 'New York'}
```

update() method:

```python
my_dict.update({'city': 'Los Angeles', 'job': 'Engineer'})  # Update existing and add new
print(my_dict)  # Output: {'name': 'Alice', 'age': 31, 'city': 'Los Angeles', 'job': 'Engineer'}
```

Deleting Items
del keyword:

```python
del my_dict['city']  # Delete a specific key-value pair
print(my_dict)  # Output: {'name': 'Alice', 'age': 31, 'job': 'Engineer'}
```

pop() method:


```python
removed_value = my_dict.pop('job')  # Remove and return the value of a key
print(my_dict)  # Output: {'name': 'Alice', 'age': 31}
print(removed_value)  # Output: Engineer
```

popitem() method:

```python
removed_item = my_dict.popitem()  # Remove and return a random key-value pair
print(my_dict)  # Output: {'name': 'Alice'}
print(removed_item)  # Output: ('age', 31)
```

clear() method:

```python
my_dict.clear()  # Remove all items from the dictionary
print(my_dict)  # Output: {}
```

Note: Dictionaries are mutable, so changes made to a dictionary directly affect the original dictionary.

**Q. Discuss the importance of dictionary keys being immutable and provide examples.**

Ans.Sure, dictionary keys are important to be immutable because they are used for lookups in the dictionary. If a key were mutable, it could be modified after it was used to add a value to the dictionary. This could lead to unexpected behavior, as the key would no longer match the value that was associated with it.

Here's an example to illustrate the importance of immutability for dictionary keys:



```python
# Mutable key (list) causes unexpected behavior
person = {'name': ['Alice', 'Bob']}  # Mutable key (not recommended)
person['name'].append('Charlie')  # Modifying the list mutates the dictionary
print(person)  # Output: {'name': ['Alice', 'Bob', 'Charlie']}

# Immutable key (tuple) ensures consistent behavior
person = {'name': ('Alice', 'Bob')}  # Immutable key (recommended)
# person['name'].append('Charlie')  # This would cause an error (tuples are immutable)
print(person)  # Output: {'name': ('Alice', 'Bob')}
```

In the first example, we create a dictionary with a mutable key ('name'). The value associated with this key is a list containing the names "Alice" and "Bob". After the dictionary is created, we modify the list by appending the name "Charlie" to it. This modification also mutates the dictionary, as the list is the key that is used to look up the value. As a result, the dictionary now contains a key-value pair where the key is the list ['Alice', 'Bob', 'Charlie'] and the value is the list ['Alice', 'Bob', 'Charlie'].

In the second example, we create a dictionary with an immutable key ('name'). The value associated with this key is a tuple containing the names "Alice" and "Bob". Tuples are immutable, so they cannot be modified after they are created. This means that the key in the dictionary cannot be changed after it is added. As a result, the dictionary remains consistent, and the key-value pair remains the same.

As you can see from this example, it is important for dictionary keys to be immutable. If keys were mutable, it could lead to unexpected behavior and make it difficult to reason about the contents of a dictionary. By using immutable keys, you can ensure that your dictionaries are more reliable and easier to work with.
